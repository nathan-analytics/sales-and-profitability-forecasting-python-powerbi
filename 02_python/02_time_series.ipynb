# Time Series Transformation

Prepare monthly sales and profit data for trend analysis and forecasting.

## Load Validated Dataset
Load the cleaned and validated dataset for time series transformation.

```python
import pandas as pd
```

```python
df = pd.read_csv("D:/Projects/super-store-orders-forecasting/01_data_source/superstore_cleaned.csv")
df["order_date"] = pd.to_datetime(df["order_date"])
df.head()
```

## Review Date Coverage
Confirm the dataset spans an appropriate time range for monthly trend analysis.

```python
df["order_date"].min(), df["order_date"].max()
```

## Aggregate Monthly Sales and Profit
Group transaction-level data by month to create a time series dataset for analysis.

```python
monthly_sales = (
    df.groupby(pd.Grouper(key="order_date", freq="ME"))
    .agg(
        total_sales=("sales", "sum"),
        total_profit=("profit", "sum")
    )
    .reset_index()
)

monthly_sales.head()
```

## Sort Monthly Data
Ensure the monthly dataset is ordered correctly for trend calculations.

```python
monthly_sales = monthly_sales.sort_values("order_date").reset_index(drop=True)
monthly_sales.head()
```

## Create Rolling Averages
Calculate rolling averages to smooth short-term fluctuations and highlight longer-term trends.

```python
monthly_sales["rolling_sales_3mo"] = monthly_sales["total_sales"].rolling(window=3).mean()
monthly_sales["rolling_profit_3mo"] = monthly_sales["total_profit"].rolling(window=3).mean()

monthly_sales.head(10)
```

## Review Time Series Output
Inspect the final monthly dataset before using it for visualization and forecasting.

```python
monthly_sales.info()
monthly_sales.describe()
monthly_sales.head(12)
```

## Export Monthly Dataset
Save the transformed monthly dataset for forecasting and dashboard development.

```python
monthly_sales.to_csv("D:/Projects/super-store-orders-forecasting/03_analysis_output/monthly_sales_timeseries.csv", index=False)
```
